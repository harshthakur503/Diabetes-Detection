```python
import pandas as pd

def compare_files(file1, file2):
  
  df1 = pd.read_csv(file1, skiprows=3, usecols=list(range(4, 13)))
  df2 = pd.read_csv(file2, skiprows=3, usecols=list(range(4, 13)))

  # Round the values in both DataFrames to integers.

  df1 = df1.round(0)
  df2 = df2.round(0)

  # Compare the values in each row for both DataFrames.

  diff_df = pd.DataFrame(columns=['Row', 'Column', 'File1', 'File2'])
  for row in range(df1.shape[0]):
    for col in range(df1.shape[1]):
      if df1.iloc[row, col] != df2.iloc[row, col]:
        diff_df = diff_df.append({
            'Row': row + 1,
            'Column': col + 1,
            'File1': df1.iloc[row, col],
            'File2': df2.iloc[row, col],
        }, ignore_index=True)

  return diff_df

if __name__ == '__main__':
  file1 = r'C:\Users\hthakur2\OneDrive - Teck Resources Limited\Documents\Automation for BASE METALS\AutoSource.csv'
  file2 = r"C:\Users\hthakur2\OneDrive - Teck Resources Limited\Documents\Automation for BASE METALS\AutoTarget.csv"

  diff_df = compare_files(file1, file2)

  print(diff_df.to_string())
```

```python
import csv
import itertools

def compare_csv_files(file1, file2):

  reader1 = csv.reader(open(file1, 'r'), delimiter=',')
  reader2 = csv.reader(open(file2, 'r'), delimiter=',')

  # Skip the header rows in both CSV files.

  next(reader1, None)
  next(reader2, None)

  n_rows = len(list(reader1)) - 1
  if n_rows == 0:
    return pd.DataFrame()

  diff_df = pd.DataFrame(columns=['Row', 'Common Group', 'Column', 'File1', 'File2'])
  for row in itertools.count(1):
    if row > n_rows:
      break
    values1 = list(reader1)[row - 1]
    values2 = list(reader2)[row - 1]

    # Check if the Common Group values are equal.

    if values1[0] != values2[0]:
      continue

    # Compare the remaining columns.

    for col in range(3, min(len(values1), 13)):
      val1 = round(float(values1[col]), 0)
      val2 = round(float(values2[col]), 0)
      if val1 != val2:
        diff_df = diff_df.append({
            'Row': row,
            'Common Group': values1[0],
            'Column': col,
            'File1': val1,
            'File2': val2,
        }, ignore_index=True)

  return diff_df

if __name__ == '__main__':
  file1 = r"C:\Users\hthakur2\OneDrive - Teck Resources Limited\Documents\Automation for BASE METALS\AutoSource.csv"
  file2 = r"C:\Users\hthakur2\OneDrive - Teck Resources Limited\Documents\Automation for BASE METALS\AutoTarget.csv"

  diff_df = compare_csv_files(file1, file2)

  print(diff_df.to_string())
```
